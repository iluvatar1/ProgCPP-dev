# Example of a jupyter notebook running c++ code

Here we mix the jupyter notebook and shell commands to develop c++ inside a notebook. Notice that there are other approaches, like https://github.com/jupyter-xeus/xeus-cling , but here we want to keed the traditional c++ developing mode. 

The next cell will create a cPP file, and then it will be compiled and executed:

```python
%%writefile helloworld_from_jupyter.cpp

#include <iostream> 

int main(int argc, char **argv)
{
    std::cout << "Hello World!" << std::endl;
    return 0;
}
```

Now let's compile it

```python
!g++ -fsanitize=address helloworld_from_jupyter.cpp
```

And now let's execute it

```python
!./a.out
```
